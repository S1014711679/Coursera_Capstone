Importing libraries

```python
import pandas as pd
```

```python
toronto = pd.read_csv('Toronto.csv',index_col = 'Postal Code')
toronto.head()
```

Adding two columns 
- Latitude 
- Longitude

```python
toronto["Latitude"] = ""
toronto["Longitude"] = ""
toronto.head()
```

```python
coords = pd.read_csv('Geospatial_Coordinates.csv',index_col='Postal Code')
coords.head()
```

Adding latitude and longitude data to Toronto dataframe based on 'postal code'

```python
print(coords.index)
print(toronto.index)
```

```python
for (index,postal_code) in enumerate(coords.index):
#     print(postal_code)
    toronto.loc[postal_code]['Latitude'] = coords.loc[postal_code]['Latitude']
    toronto.loc[postal_code]['Longitude'] = coords.loc[postal_code]['Longitude']
```

Saving the data to Toronto_Map.csv

```python
toronto.to_csv('Toronto_Map.csv')
```
